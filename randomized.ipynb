```python
import random

# Set a random seed for reproducibility
random.seed(42)

# Define your data in a list of tuples (category, case)
data = [
    ("common malignant salivary", 1),
    ("common malignant salivary", 2),
    ("common malignant salivary", 3),
    ("rare malignant salivary", 4),
    ("rare malignant salivary", 5),
    ("rare malignant salivary", 6),
    ("common benign salivary", 7),
    ("common benign salivary", 8),
    ("common benign salivary", 9),
    ("rare benign salivary", 10),
    ("rare benign salivary", 11),
    ("common odontogenic tumor", 12),
    ("common odontogenic tumor", 13),
    ("common odontogenic tumor", 14),
    ("rare odontogenic tumor", 16),
    ("common bone tumor", 17),
    ("common bone tumor", 18),
    ("common bone tumor", 21),
    ("common odontogenic cyst", 22),
    ("common odontogenic cyst", 23),
    ("common odontogenic cyst", 24),
    ("common odontogenic cyst", 25),
    ("rare odontogenic tumor", 26),
    ("rare odontogenic tumor", 27),
    ("rare bone tumor", 28),
    ("rare odontogenic cyst", 30),
    ("rare bone tumor", 31)
]

# Create a dictionary to store one randomly selected case per category
selected_cases = {}

# Shuffle the data to ensure randomness
random.shuffle(data)

# Iterate through the shuffled data and select one case per category
for category, case in data:
    if category not in selected_cases:
        selected_cases[category] = case

# Shuffle the order of categories for printing
categories_list = list(selected_cases.keys())

selection = []
# Print the selected cases in the randomized order of categories
for category in categories_list:
    case = selected_cases[category]
    selection.append(f"{category} #{case}")

common = []
rare = []
for item in selection:
    if item.startswith('c'):
        common.append(item)
    else:
        rare.append(item)

random.shuffle(common)
random.shuffle(rare)

items_per_list = 2
for i in range(0, len(common), items_per_list):
    for item in common[i:i + items_per_list]:
        print(item)
    for item in rare[i:i + items_per_list]:
        print(item)
```
